# Analysis of $F_1$ metric for a $p$-coin
## ctorres@inf.utfsm.cl

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import rc
rc('font',**{'family':'serif','serif':['Palatino'],'size':16})
rc('text', usetex=True)
```

```python
f = lambda K,p: 2*((1/(1 + K))*(1 - p))/((1/(1 + K)) + (1 - p))

K=49
fv = np.vectorize(lambda p: f(K,p))
p=np.linspace(0,1,100)
plt.figure()
plt.plot(p,fv(p),'-')
plt.grid(True)
plt.ylim([0,0.1])
plt.xlabel('$p$')
plt.ylabel('$F_{1}$')
plt.show()
```

```python
f = lambda K,p: 2* p /(p + 1 + (p*K))
K=49
fv = np.vectorize(lambda p: f(K,p))

p=np.linspace(0,1,100)
plt.figure()
plt.plot(p,fv(p),'-')
plt.grid(True)
plt.ylim([0,0.1])
plt.xlabel('probability $p$')
plt.ylabel('$F_{1}$')
plt.show()
```

```python
f = lambda K,p: 2* p /(p + 1 + (p*K))

K_range = [1,5,10,35,40,45,49,80]
plt.figure(figsize=(8,6))
#plt.figure()
for K in K_range:
    fv = np.vectorize(lambda p: f(K,p))

    p=np.linspace(0,1,100)
    if K != 49:
        plt.plot(p,fv(p),'-',alpha=0.5, label="$K$: " + str(K),linewidth=2.5,dashes=[3, 2])
    else:
        plt.plot(p,fv(p),'k-', label="$K$: " + str(K),linewidth=2.5)

    
plt.grid(True)
plt.legend(loc='upper left')
#plt.ylim([0,0.1])
plt.xlabel('probability $p$')
plt.ylabel('$F_{1}$')
plt.show()
```
