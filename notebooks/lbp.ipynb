```python
from skimage.transform import rotate
from skimage.feature import local_binary_pattern
from skimage import data
from skimage.color import label2rgb, rgb2gray,rgb2hsv,rgb2hed

from skimage import io
from matplotlib import pyplot as plt
```

```python
img_dir = "/home/rpezoa/experiment_data/"
img_type = ["2+"]
```

```python
image = io.imread(img_dir + "big_2+/img/2+_8.tif")
img = rgb2gray(image)
hsv = rgb2hsv(image[:,:,0:3])
```

```python
plt.figure(figsize=(8,8))
plt.imshow(hsv[:,:,0]<0.5,cmap="gray")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img,cmap="gray")
plt.show()
```

```python
# settings for LBP
radius = 2
n_points = 8 * radius
lbp = local_binary_pattern(img, n_points, radius, "uniform")
```

```python
plt.figure(figsize=(8,8))
plt.imshow(lbp,cmap="gray")
plt.show()
```

```python
for i in range(int(lbp.max()+1)):
    plt.figure(figsize=(10,10))
    plt.imshow(lbp==i ,cmap="gray")
    plt.title(str(i))
    plt.show()
```
