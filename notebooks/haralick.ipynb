```python
import numpy as np
from matplotlib import pyplot as plt

def show(im, c = "gray",title=""):
    plt.figure(figsize=(6,6))
    plt.imshow(im,cmap = c)
    plt.title(title)
    plt.axis("off")
    plt.show()
```
