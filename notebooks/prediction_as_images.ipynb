```python
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
images = ["1+_20", "1+_25","2+_8","2+_9","3+_19"]
type_im = ["1+","1+","2+","2+","3+"]


classifiers = ["deep_0","svm_0","rf_0", "extra_trees_0", "knn_0"]
```

```python
n_seeds = 10
n_im = len(images)
f1 = np.zeros((n_im,n_seeds))
prec = np.zeros((n_im,n_seeds))
rec = np.zeros((n_im,n_seeds))
for i in range(n_im):
    path = "/home/rpezoa/experiment_data/output/" + images[i]+ "_rpr_halton_patches/"
    f = os.listdir(path)
    #print(f)
    big_target_path="/home/rpezoa/experiment_data/big_" + type_im[i] + "/labels/" + images[i] + ".npy"
    print(":"*15 + " Im:" + images[i] + ":"*15)
    for k in classifiers:
        for j in range(n_seeds):
            big_y0=np.load(big_target_path)
            big_pred = np.load(path + k +'/big_pred/' + str(j) + '_big_pred.npy')
            im_pred = np.reshape(big_pred, (1000,1000))
            plt.figure()
            plt.imshow(im_pred,cmap="gray")
            plt.show()
            print(path + k +"/"+ str(j)+".png")
            plt.imsave(path + k +"/"+ str(j)+".png",im_pred,cmap="gray")
```
