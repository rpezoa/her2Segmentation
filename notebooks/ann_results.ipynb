```python
from matplotlib import pyplot as plt
```

```python
#batchSize=["100", "300"]
#epochs=["100", "200","500"]
epochs=["100", "300"]
batchSize=["100", "200","500"]

percentages=["0.2"]#, "0.3"]
#trainingSize=["1000", "2000", "3000", "4000", "5000", "6000"]
trainingSize=["13005"]
seeds=list(map(str,range(0,10)))

#a2 = open("F1_test.txt")
#a3 = open("F1_big.txt")

def read_file(base_dir,file_name):
    d={}
    #d2={}
    #d3={}

    for bs in batchSize:
        for e in epochs:
            for p in percentages:
                for s in seeds:

                    preffix = "_".join([str(bs),str(e),str(p),s,""])
                    a1 = open(base_dir +preffix + file_name )
                    for l1 in a1:
                        lista1=l1.strip().split(":")
                        #print(lista1)
                        trainSize,se,f1 = lista1[0],lista1[1],lista1[2]
                        if trainSize not in d:
                            d[trainSize] = []
                        else:
                            d[trainSize].append((f1,bs,e,p,se))
                    a1.close()
    return d
```

```python
def plot_performance(l,ylabel,xlabel,mainTitle, title,percentage_test,out_dir="out/plots/svm/",training_type="strati"):
    fig, axes = plt.subplots(figsize=(18, 6), nrows=1, ncols=3)
    fig.suptitle(mainTitle + ", " + ylabel+ ", perc. test: " +", " + percentage_test, fontsize=16)

    titles = ["Training","Testing","Big image"]
    
    Y=[]
    for i,dic in enumerate(l):
        score = {}
        for ts,datos in dic.items():
            for sc, bs,e,p,se in datos:
                if p == percentage_test:
                    if ts not in score:
                        score[ts] = []
                    score[ts].append(float(sc))
        x,y=[],[]

        
        for ts in score:
            Y.append(max(score[ts]))
            x.append(ts)
            y.append(max(score[ts]))
        print("-->",x,y)
    
    
    
        axes[i].set_title(titles[i])
        axes[i].plot(x,y,'o')
        axes[i].set_xlabel(xlabel)
        axes[i].set_ylabel(ylabel)
        axes[i].set_ylim(0,1)
        axes[i].grid()
        
        plt.savefig(out_dir + training_type + ylabel+ mainTitle + "_" + percentage_test + ".png", bbox_inches="tight")
    return Y
```

```python
images = ["1+_20", "1+_25", "2+_1","2+_8","2+_9","3+_19"]
images = ["3+_19"]
for im in images:

    #base_dir = "/home/rpezoa/her2Segmentation/notebooks/out/"+ im +"/ann_patches/"
    base_dir = "/home/rpezoa/her2Segmentation/notebooks/out/"+ im +"/ann_strati/"
    d1 = read_file(base_dir,"F1_train.txt")
    d2 = read_file(base_dir,"F1_test.txt")
    d3 = read_file(base_dir,"F1_big.txt")

    d4 = read_file(base_dir,"PREC_train.txt")
    d5 = read_file(base_dir,"PREC_test.txt")
    d6 = read_file(base_dir,"PREC_big.txt")


    d7 = read_file(base_dir,"REC_train.txt")
    d8 = read_file(base_dir,"REC_test.txt")
    d9 = read_file(base_dir,"REC_big.txt")
    
    #print(d1)
    
    y1 = plot_performance([d1,d2,d3],"F1","training size","Image: "+ im, "Training data",'0.2' )
    y2 = plot_performance([d4,d5,d6],"Precision","training size","Image: "+ im, "Training data",'0.2' )
    y3 = plot_performance([d7,d8,d9],"Recall","training size","Image: "+ im, "Training data",'0.2' )
```

```python
def plot_f1(d1,ylabel,xlabel,title,percentage_test):
    f1_train = {}
    for ts,datos in d1.items():
        for f1, bs,e,p in datos:
            if p == '0.2':
                if ts not in f1_train:
                    f1_train[ts] = []
                f1_train[ts].append(float(f1))

    x,y=[],[]

    for ts in f1_train:
        x.append(ts)
        y.append(max(f1_train[ts]))
    #print(x, y)
    
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    #x = np.linspace(1., 8., 30)
    #ax.plot(x, x ** 1.5, 'ro', label='density')
    #ax.plot(x, 20/x, 'bx', label='temperature')
    
    #plt.figure()
    ax.plot(x,y,'o')
    ax.set_ylim(0,1)
    ax.set_title(title + "perc. test: " +", " + percentage_test)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid()
    #ax.show()
    return ax
```
